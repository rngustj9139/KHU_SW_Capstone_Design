```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
train_learning_rate_list = [] # Learning Rate
train_accuracy_list = [] # Accuracy
train_loss_cls_list = [] # Classification Loss
train_loss_bbox_list = [] # Bbox Regression Loss
train_loss_list = [] # Total Loss
train_loss_rpn_cls_list = [] # RPN Classification Loss
train_loss_rpn_bbox_list = [] # RPN Bbox Regression Loss
```

```python
val_learning_rate_list = [] # Learning Rate
val_ap50_list = [] # AP50 (IOU50 적용시 Average Precision)
val_map_list = [] # mAP
```

```python
import json

def getInformationsFromLogFile(filePath):
  with open(filePath) as file:
    oneJsonFile = json.load(file)

    for oneJsonData in oneJsonFile:
      if (oneJsonData["mode"] == 'train'):
        train_learning_rate_list.append(oneJsonData["lr"]) # Learning Rate
        train_accuracy_list.append(oneJsonData["acc"]) # Accuracy
        train_loss_cls_list.append(oneJsonData["loss_cls"]) # Classification Loss
        train_loss_bbox_list.append(oneJsonData["loss_bbox"]) # Bbox Regression Loss
        train_loss_list.append(oneJsonData["loss"]) # Total Loss
        train_loss_rpn_cls_list.append(oneJsonData["loss_rpn_cls"]) # RPN Classification Loss
        train_loss_rpn_bbox_list.append(oneJsonData["loss_rpn_bbox"]) # RPN Bbox Regression Loss

      elif (oneJsonData["mode"] == 'val'):
        val_learning_rate_list.append(oneJsonData['lr']) # Learning Rate
        val_ap50_list.append(oneJsonData['AP50']) # AP50 (IOU50 적용시 Average Precision)
        val_map_list.append(oneJsonData['mAP']) # mAP

    return train_learning_rate_list, train_accuracy_list, train_loss_cls_list, train_loss_bbox_list, train_loss_list, train_loss_rpn_cls_list, train_loss_rpn_bbox_list, val_learning_rate_list, val_ap50_list, val_map_list
```
