```python
  train_learning_rate_list = []
  train_triplet_loss_list = []
  train_ce_loss_list = []
  train_top_1_accuracy_list = []
  train_total_loss_list = []
```

```python
import json

def getInformationsFromLogFile(filePath):
  with open(filePath) as file:
    oneJsonFile = json.load(file)

    for oneJsonData in oneJsonFile:
      train_learning_rate_list.append(oneJsonData["lr"])
      train_triplet_loss_list.append(oneJsonData["triplet_loss"])
      train_ce_loss_list.append(oneJsonData["ce_loss"])
      train_top_1_accuracy_list.append(oneJsonData["top-1"])
      train_total_loss_list.append(oneJsonData["loss"])

  return train_learning_rate_list, train_triplet_loss_list, train_ce_loss_list, train_top_1_accuracy_list, train_total_loss_list
```
